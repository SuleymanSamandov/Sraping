```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def scrapumico():
    list_ = []
    baseurl = 'https://umico.az'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
    for x in range(1,2):
        url = f'https://umico.az/categories/3-mobil-telefonlar-ve-smartfonlar?page={x}'
        try:
            r = requests.get(url, headers=headers)
            if r.status_code == 200:
                soup = BeautifulSoup(r.content, 'html.parser')
                productlist = soup.find_all('div', class_='MPProductItem')  

                for item in productlist:
                    for link in item.find_all('a', href=True):
                        list_.append(baseurl + link['href'])
            else:
                print("Wrong status code:", r.status_code)

        except requests.exceptions.RequestException as e:
            print('An error occurred:', e)
            
    parfumlist=[]
    for link in list_:
        r = requests.get(link)

        try:
            if r.status_code == 200:
                soup = BeautifulSoup(r.content, 'html.parser')
                name = soup.find('h1', itemprop="name")
                if name:
                    name = name.get_text(strip=True)
                
                originalprice=soup.find("span",class_="MPPrice-OldPrice")
                if originalprice:
                    originalprice = originalprice.get_text(strip=True)
                
                
                brend=soup.find("span",class_="v")
                if brend:
                    brend = brend.get_text(strip=True)
                
                device=soup.find("div",class_="v")
                if device:
                    device = device.get_text(strip=True)

               
                    parfum = {
                        "Name": name,
                        "Originalprice":originalprice,
                        "Brend":brend,
                        "Device":device
                        
                    }
                    parfumlist.append(parfum)
                
        except requests.exceptions.RequestException as e:
            print('An error occurred:', e)
            
        except Exception as e:
            print('An unexpected error occurred:', e)
    df=pd.DataFrame(parfumlist)        
    return df
```

```python
scrapumico()
```
